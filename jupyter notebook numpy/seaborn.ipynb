### boxplot

```python
### Box plot in seaborn

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Create sample data
data = sns.load_dataset('iris')
# data = pd.DataFrame({
#     'Category': ['A']*50 + ['B']*50,
#     'Value': np.random.normal(0, 1, 100)
# })
# features= data.feature_names
# print(features)
sns.boxplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Box Plot')
plt.show()
```

### Violin Plot

```python
# data = pd.DataFrame({
#     'Category': ['A']*50 + ['B']*50,
#     'Value': np.random.normal(0, 1, 100)
# })
sns.violinplot(x='sepal_width', y='sepal_length' , data=data, color='green')
plt.title("violinplot")
plt.show()
```

### Histogram

```python
sns.histplot(data=data, x='sepal_width',bins=30, kde=True)
plt.title('Histogram with KDE')
plt.show()
```

### KDE plot

```python
sns.kdeplot(data=data, x='Value', hue='Category')
plt.title('Kernel Density Estimation')
plt.show()
```

### Bar plot

```python
sns.barplot(x='Category', y='Value', hue='Category', data=data)
plt.title('Seaborn Bar Plot')
plt.show()
```

### count plot

```python
sns.countplot(x='Category', data=data, hue='Category')
plt.title('Count Plot')
plt.show()
```

### heatmap

```python
## heatmap in seaborn
data = data.apply(pd.to_numeric, errors='coerce')
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## pair plot

```python
# Create sample data
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

### joint plot

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter')
plt.show()
```

### FAced Grid

```python
g = sns.FacetGrid(iris, col='species')
g.map(sns.histplot, 'sepal_length')
plt.show()
```
